#Obtençao de Dados

```
ibov = web.get_data_yahoo("^BVSP")
#ibov = web.get_data_yahoo("^BVSP", start = '2010-05-03', end = '2013-03-01')
```

```
ibov.head()
```

```
ibov.tail()
```

```
ibov["Close"].plot(figsize=(22,8), label= "IBOV")
ibov["Close"].rolling(200).mean().plot(label="MM200")
ibov["Close"].rolling(21).mean().plot(label="MM21")
ibov["Close"].rolling(9).mean().plot(label="MM9")
plt.legend()
```

```
#Definindo um intevalo menor
ibov_fatiado = ibov[ibov.index.year == 2008]
ibov_fatiado["Close"].plot(figsize=(22,8), label = "IBOV")
ibov_fatiado["Close"].rolling(9).mean().plot(label="MM9")
ibov_fatiado["Close"].rolling(21).mean().plot(label="MM21")
plt.legend()
```

```
#Definindo outro ano
ibov_fatiado = ibov[ibov.index.year == 2019]
ibov_fatiado["Close"].plot(figsize=(22,8), label = "IBOV")
ibov_fatiado["Close"].rolling(9).mean().plot(label="MM9")
ibov_fatiado["Close"].rolling(21).mean().plot(label="MM21")
plt.legend()
```

```
#Definindo um intevalo 
ibov_fatiado = ibov[(ibov.index.year >= 2000) & (ibov.index.year <=2009)]
ibov_fatiado["Close"].plot(figsize=(22,8), label = "IBOV")
ibov_fatiado["Close"].rolling(9).mean().plot(label="MM9")
ibov_fatiado["Close"].rolling(21).mean().plot(label="MM21")
plt.legend()
```

```
#Definindo um intevalo 
ibov_fatiado = ibov[(ibov.index.year >= 2016) & (ibov.index.year <=2020)]
ibov_fatiado["Close"].plot(figsize=(22,8), label = "IBOV")
ibov_fatiado["Close"].rolling(9).mean().plot(label="MM9")
ibov_fatiado["Close"].rolling(21).mean().plot(label="MM21")
plt.legend()
```
